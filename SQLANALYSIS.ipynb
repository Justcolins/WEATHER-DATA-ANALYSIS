```python
import sqlite3
import pandas as pd
```

```python
df = pd.read_csv("F:\\WEATHER ANALYSIS\\1. Weather Data.csv")
```

```python
conn = sqlite3.connect('weather.db')
```

```python
df.to_sql('weather', conn, index=False, if_exists='replace')
```

# 1. RECORDS OF CLEAR WEATHER

```python
query = """
SELECT *
FROM Weather
WHERE Weather = 'Clear'
"""
result = pd.read_sql_query(query, conn)
result
```

# 2. WIND SPEED EQUAL TO 4KM/HR

```python
query2 = """
SELECT * 
FROM Weather
WHERE "Wind Speed_km/h" = 4;
"""
result2 = pd.read_sql_query(query2,conn)
result2
```

# 3. WIND SPEED > 24KM/HR AND VISIBILITY =25

```python
query3 = """
SELECT *
FROM Weather
WHERE "Wind Speed_km/h" > 24 
AND "Visibility_km" = 25;
"""
result3 = pd.read_sql_query(query3,conn)
result3
```

# 4. MEAN VISIBILITY

```python
query4 = """
SELECT AVG(Visibility_km) AS Mean_Visibility
FROM Weather

"""
result4 = pd.read_sql_query(query4,conn)
result4
```
